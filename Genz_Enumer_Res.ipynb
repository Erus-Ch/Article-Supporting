```python
# This program is written to enumerate the symmetric adjacency matrices for generalized catenanes;
# where N is the number of monomers, and M-1 is the number of cell (cavity) in each monomer;
# which is equivlanent ot enumerate (M-1)-Dyck paths and transform the path into adjacency matrices of our catenanes;
# The algorithm is based on our results (proposition, definition and theorem) in article;
import itertools
import numpy


def comb(n: int, r: int)-> list:
    if n == r:
        return [[i for i in range(1,n+1)]]
    if r == 1:
        return [[i] for i in range(1,n+1)]
    ans = []
    for each_list in comb(n-1,r-1):
        ans.append([n]+each_list)
    for each_list in comb(n-1,r):
        ans.append(each_list)
    return ans
#k-Dyck path
M=4 # (M-1) is the number of cavities in any monomer;
N=6 # N is the numerb of monomers in catenanes (degree of polymerization)
tmp=[]
Dpath=[]
tmp=comb(M*N,N)
#print(tmp)
for i in range(0,len(tmp)):
    chk=[]
    chk=tmp[i]
    chk.sort()
    #print(chk)
    rs=0
    for j in range(0,N-1):
        if ((chk[j]-(j+1)*M)>0):
            rs=rs+1
    if (rs==N-1):
        if (chk[N-1]==M*N):
            Dpath.append(chk)

#print(Dpath)
print(len(Dpath))
LM=[]
for l in range(0,len(Dpath)):
    seq=Dpath[l]
    for i5 in range(0,N):
        seq[i5]=seq[i5]-(i5+1)*M
    AdM=[]
    for i0 in range(0,N):
        tx=[]
        for j0 in range(0,N):
            tx.append(0)
        AdM.append(tx)
    #print(seq)
    AdM[N-2][N-1]=seq[N-2]
    for i in range(N-3,-1,-1):
        p1=seq[i]
        p2=seq[i+1]
        for j in range(N-1,i+1,-1):
            AdM[i][j]=p1-(j-i-1)*(M-1)
            for r in range(i+1,j):
                AdM[i][j]=AdM[i][j]+AdM[r][j]
            AdM[i][j]=AdM[i][j]//(j-i)
            if (AdM[i][j]<0):
                AdM[i][j]=0
            p1=p1-AdM[i][j]
            p2=p2-AdM[i+1][j]
        AdM[i][i+1]=p1
    #print(AdM)
    pr=0
    for i1 in range(0,N):
        for j1 in range(i1+1,N):
            if (AdM[i1][j1]==AdM[N-j1-1][N-1-i1]):
                pr=pr+1
            pr=pr-1
    if (pr==0):
        LM.append(AdM)

print(len(LM),LM)


    
```

```python
# This program is written to enumerate the valid adjacency matrices for generalized catenanes underestriction of finite cavity volume.
# where N is the number of monomers, and M-1 is the number of cell (cavity) in each monomer
# p is used to denote the relative volume for containing occupied monomers;
# which is also equivlanent ot enumerate (M-1)-Dyck paths and transform the path into adjacency matrices of our catenanes
# The algorithm is based on our results (proposition, definition and theorem) in article;
import itertools
import numpy


def comb(n: int, r: int)-> list:
    if n == r:
        return [[i for i in range(1,n+1)]]
    if r == 1:
        return [[i] for i in range(1,n+1)]
    ans = []
    for each_list in comb(n-1,r-1):
        ans.append([n]+each_list)
    for each_list in comb(n-1,r):
        ans.append(each_list)
    return ans
#k-Dyck path
M=3 # (M-1) is the number of cavities in any monomer;
N=4 # a positive integer "N" is the numerb of monomers in catenanes (degree of polymerization)
p=3 # a positive integer "p" is used to denote the relative volume for containing occupied monomers;
tmp=[]
Dpath=[]
tmp=comb(M*N,N)
#print(tmp)
for i in range(0,len(tmp)):
    chk=[]
    chk=tmp[i]
    chk.sort()
    #print(chk)
    rs=0
    for j in range(0,N-1):
        if ((chk[j]-(j+1)*M)>0):
            rs=rs+1
    if (rs==N-1):
        if (chk[N-1]==M*N):
            Dpath.append(chk)

#print(Dpath)
print(len(Dpath))
LM=[]
for l in range(0,len(Dpath)):
    seq=Dpath[l]
    for i5 in range(0,N):
        seq[i5]=seq[i5]-(i5+1)*M
    AdM=[]
    for i0 in range(0,N):
        tx=[]
        for j0 in range(0,N):
            tx.append(0)
        AdM.append(tx)
    #print(seq)
    AdM[N-2][N-1]=seq[N-2]
    for i in range(N-3,-1,-1):
        p1=seq[i]
        p2=seq[i+1]
        for j in range(N-1,i+1,-1):
            AdM[i][j]=p1-(j-i-1)*(M-1)
            for r in range(i+1,j):
                AdM[i][j]=AdM[i][j]+AdM[r][j]
            AdM[i][j]=AdM[i][j]//(j-i)
            if (AdM[i][j]<0):
                AdM[i][j]=0
            p1=p1-AdM[i][j]
            p2=p2-AdM[i+1][j]
        AdM[i][i+1]=p1
    #print(AdM)
    pr=0
    for i1 in range(0,N): # The i1-th monomer
        for j1 in range(1,M): #The j1-th cell cavity
            q=0
            for k1 in range(0,i1):
                if ((AdM[k1][i1]-j1+1)>0):
                    q=q+1
            for k2 in range(i1+1,N):
                if ((AdM[i1][k2]-M+1+j1)>0):
                    q=q+1
            if (q>p):
                pr=pr+1
    if (pr==0): # Only if all cell filling is valid under restriction of "p"
        LM.append(AdM)

print(len(LM),LM)
```
